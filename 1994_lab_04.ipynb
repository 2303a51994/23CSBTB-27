```python
import matplotlib.pyplot as plt
import numpy as np

xpoints = np.array([0,1,3,4,6])
ypoints = np.array([0,100,200,50,250])

plt.plot(xpoints,ypoints)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

ypoints = np.array([3, 8, 1, 10, 5, 3, 7])

plt.plot(ypoints ,"o:b")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

ypoints=np.array([3,41,3,105,7])
plt.plot(ypoints,marker='o',ms=20,mec='red',mfc='yellow')
plt.plot(ypoints,ls=':')
plt.plot(ypoints,color='r')
plt.plot(ypoints,linewidth='20.5')

plt.plot(ypoints, "p:y")
plt.show()
```
